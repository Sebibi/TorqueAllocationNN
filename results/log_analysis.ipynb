```python
import pandas as pd
```

```python
# Load the data
file_name = 'TA_optimization.csv'
df = pd.read_csv(file_name, index_col=0)
df = df.iloc[:]
df.head()
```

```python
df.describe()
```

```python
# PLot the distribution of the data
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

fig, ax = plt.subplots(figsize=(12, 6))
sns.histplot(df.iloc[:, :6], kde=True)
plt.title('Distribution of the input data')
plt.show()
```

```python
# Plot the distribution of the output data
fig, ax = plt.subplots(figsize=(12, 6))
sns.histplot(df.iloc[:, 6:10], kde=True)
plt.title('Distribution of the output data')
plt.show()
```
